# LSA

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from nltk.tokenize.casual import casual_tokenize
from sklearn.decomposition import PCA
from sklearn.decomposition import TruncatedSVD
from gensim.utils import tokenize
```

```python
def clean_text(text):
    text=re.sub(r"[^a-zA-Z0-0]"," ",text)
    text =' '.join(text.split())
    return text.lower()
```

```python
test = list(map(clean_text,i_contents[281]))
```

```python
test
```

```python
# tfidf = TfidfVectorizer(tokenizer=casual_tokenize,stop_words=stopwords,max_features=1000,max_df=0.5,smooth_idf=True)
tfidf = TfidfVectorizer(tokenizer=casual_tokenize,stop_words=stopwords,max_features=100,max_df=0.5,smooth_idf=True)
tfidf_docs = tfidf.fit_transform(raw_documents=test).toarray()
```

```python
tfidf
```

```python
tfidf_docs = pd.DataFrame(tfidf_docs)
tfidf_docs - tfidf_docs.mean()
```

```python
len(tfidf_docs)
```

```python
pca = PCA(n_components=len(tfidf_docs)//4)
pca = pca.fit(tfidf_docs)
pca_topic_vectors = pca.transform(tfidf_docs)
col = ['topic{}'.format(i) for i in range(pca.n_components)]
pca_topic_vectors = pd.DataFrame(pca_topic_vectors, columns=col)
pca_topic_vectors.round(3).head(10)
```

```python
tfidf.vocabulary_
colum_nums, terms = zip(*sorted(zip(tfidf.vocabulary_.values(), tfidf.vocabulary_.keys())))
```

```python
weights = pd.DataFrame(pca.components_, columns=terms,index=['topic{}'.format(i) for i in range(len(tfidf_docs)//4)])
weights.head(5).round(3)
```

```python
weights.iloc[9].sort_values(ascending=False).head(50)
```

```python
t = weights['science fiction thriller'.split()].round(3)*100
t
```

```python
svd = TruncatedSVD(n_components=len(tfidf_docs)//2, n_iter=100)
svd_topic_vectors = svd.fit_transform(tfidf_docs.values)
svd_topic_vectors = pd.DataFrame(svd_topic_vectors)
svd_topic_vectors.round(3).head(5)
```

```python
import numpy as np
svd_topic_vectors = (svd_topic_vectors.T / np.linalg.norm(svd_topic_vectors,axis=1)).T
```

```python
svd_topic_vectors.iloc[:].dot(svd_topic_vectors.iloc[:].T).round(2)
```
